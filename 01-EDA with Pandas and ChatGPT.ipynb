```python
import pandas as pd

# Assuming the file is in the same directory as the notebook
file_path = "air_pollution.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)


```

```python
# Display the first few rows of the DataFrame to verify the data is loaded correctly
df.head()
```

```python
column_names = df.columns
print(column_names)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Get summary statistics of numerical columns
df.describe()


```

```python
# Check the number of rows and columns in the dataset
df.shape
```

```python

# Check for missing values in the dataset
df.isnull().sum()
```

```python
# Increase the figure size
plt.figure(figsize=(10, 6))

# Customize the histogram appearance
sns.histplot(data=df, x='pollution_today', kde=True)

# Set labels and title
plt.xlabel('Pollution Today')
plt.ylabel('Frequency')
plt.title('Distribution of Pollution Today')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```

```python
# Increase the figure size
plt.figure(figsize=(10, 6))

# Customize the scatter plot appearance
sns.scatterplot(data=df, x='temp', y='pollution_today')

# Set labels and title
plt.xlabel('Temperature')
plt.ylabel('Pollution Today')
plt.title('Temperature vs. Pollution Today')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```

```python
# Increase the figure size
plt.figure(figsize=(10, 8))

# Calculate the correlation matrix
corr_matrix = df.corr()

# Customize the heatmap appearance
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Set title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```

```python
# Grouping by date and calculating mean pollution_today
daily_mean_pollution = df.groupby('date')['pollution_today'].mean()
# Increase the figure size
plt.figure(figsize=(12, 6))

# Line plot of mean pollution_today over time
daily_mean_pollution.plot()

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Mean Pollution Today')
plt.title('Mean Pollution Today Over Time')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```
