```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('01-air_pollution.csv')
```

```python
# Create box plots for each variable
plt.figure(figsize=(12, 6))
df.iloc[:, 1:].boxplot(grid=False)
plt.xticks(rotation=45)
plt.title('Box plots - Outlier Identification')
plt.show()
```

```python
from scipy import stats

z_scores = stats.zscore(df.iloc[:, 1:])  # Exclude the 'date' column
```

```python
threshold = 3  # Set the threshold for outlier detection

# Replace outliers with mean
outliers = np.abs(z_scores) > threshold
df.iloc[:, 1:] = np.where(outliers, np.nanmean(df.iloc[:, 1:], axis=0), df.iloc[:, 1:])
```

```python
# Create box plots for each variable after replacing outliers
plt.figure(figsize=(12, 6))
df.iloc[:, 1:].boxplot(grid=False)
plt.xticks(rotation=45)
plt.title('Box plots after Outlier Replacement')
plt.show()

```
